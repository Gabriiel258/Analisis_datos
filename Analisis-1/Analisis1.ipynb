```python
#obtener el dominio de la columba email
```

```python
def obtener_dominio(email):
    dominio = email.split('@')[-1].replace('www.', '')
    return dominio
```

```python
df_universidades['Dominio'] = df_universidades['E_Mail'].apply(obtener_dominio)
```
